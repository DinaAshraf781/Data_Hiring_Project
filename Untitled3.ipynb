```python
import pandas as pd
import numpy


logs = pd.read_json('/content/logs.json')
logs.to_excel('Logs.xlsx')

app = pd.read_excel('/content/Applicants.xlsx')

app['fit_to_account'] = app['fit_to_account'].replace('[]','none')

for i in app:
  x = app[i].isnull().sum()
  if x !=0 and app[i].dtype == object:
    print(x,app[i].name)
    app[i] = app[i].fillna('none')
  else:
    app[i] = app[i].replace(0,'')
app
```
